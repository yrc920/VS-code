```python
import numpy as np
import pandas as pd
```

```python
series = pd.Series(np.random.randint(0,100, (5, )), index = ['a', 'b', 'c', 'd', 'e'])
series
```

```python
# 1. 行索引
# series['a']
series[0]
```

```python
# 2. 切片索引（取到的值都是连续的）
# 名称：是会包含末尾的值
# 下标：不会包含末尾的值
# series['a': 'c']
series[0:3]
```

```python
# 3. 不连续索引（要获取多个）
series[['a', 'c', 'd']]
```

```python
# 4. 条件索引
# condition = series > 50
# series[condition]

series[series>50]
```
