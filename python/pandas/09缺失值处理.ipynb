```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(
    np.random.randn(5, 3),
    index=["a", "c", "e", "f", "h"],
    columns=["one", "two", "three"],
)

df['four'] = 'bar'
df['five'] = df['one'] > 0

df = df.reindex(["a", "b", "c", "d", "e", "f", "g", "h"])
df
```

```python
# 1. 判断是否有空值
np.any(df.isna())

np.all(df.notna())
# df.info()
```

```python
# 2. 填充NAN
df['two'].fillna(0)
```

```python
df['two'].fillna(df['two'].mean())
```

```python
# 3. 删除NAN
df.dropna(axis=0)
```

```python
df.dropna(axis=1)
```

```python
# 4. 替换
d = {"a": list(range(4)), "b": list("ab.."), "c": ["a", "b", np.nan, "d"]}
df = pd.DataFrame(d)
df
```

```python
# df.replace(".", np.NAN)
df.replace(np.NAN, "0")
```
