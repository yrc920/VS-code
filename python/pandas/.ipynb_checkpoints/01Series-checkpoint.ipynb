##### 在数据分析工作中，Pandas 的使用频率是很高的，一方面是因为 Pandas 提供的基础数据结构 DataFrame 与 json 的契合度很高，转换起来就很方便。另一方面，如果我们日常的数据清理工作不是很复杂的话，通常用几句 Pandas 代码就可以对数据进行规整。

##### pandas主要有Series 和 DataFrame 这两个核心数据结构，他们分别代表着一维的序列和二维的表结构。基于这两种数据结构，Pandas 可以对数据进行导入、清洗、处理、统计和输出。

### 数据结构：Series 和 DataFrame

##### Series 是个定长的字典序列。说是定长是因为在存储的时候，相当于两个 ndarray，这也是和字典结构最大的不同。因为在字典的结构里，元素的个数是不固定的。

###### Series 有两个基本属性：index 和 values。在 Series 结构中，index 默认是 0,1,2,……递增的整数序列，当然我们也可以自己来指定索引，比如 index=[‘a’, ‘b’, ‘c’, ‘d’]。

```python
import numpy as np
import pandas as pd
```

### Series的创建

```python
# 1. 使用列表来创建
series1 = pd.Series(['张三', '李四', '王五'])
series1
```

```python
# 2. 使用np数组
series2 = pd.Series(np.random.randint(0,99, size=(5, )))
series2
```

```python
# 3. 使用字典创建
user = {"张三": 18, "李四": 24, "王五": 36}
series3 = pd.Series(user)
series3
```

```python
# 获取Series索引值
# for value in series3.index:
#     print(value)\
series3.index.to_numpy()
```

```python
# 获取Series值
series3.values
```

```python
# series3['李四']
series3[1]
```

# 总结：

## 创建数组：
1. 根据列表或者numpy数组创建。
2. 获取Series对象的索引，是通过Series对象的index属性获取。
3. 获取Series对象的值，是通过Series对象的values属性获取。
4. 根据某个索引获取对应的值。通过series[索引名称]，也可以通过默认的下标索引获取。比如series[1]
