```python
import numpy as np
```

### 获取某行的数据：

```python
# a1 = np.arange(0,29)
# print(a1[3])
# print(a1)#获取下标为1的元素
a1 = np.arange(0,24).reshape((4,6))
print(a1[1][1])
# print(a1[1]) #获取下标为1的行的数据 
```

### 连续获取某几行的数据：

```python
a1 = np.arange(0,24).reshape((4,6))
print(a1,"\n")
print(a1[0:2]) #获取0行到1行的数据

# print(a1[[0,2,3]])

# print(a1[[-1,-4]])   #print(a1[[0,3]])
```

###  获取某行某列的数据：

```python
a1 = np.arange(0,24).reshape((4,6))
# print(a1[1,1]) #获取1行1列的数据
# print(a1[0:2,2:4]) #获取0-1行的0-1列的数据
print(a1[[1,2],[2,3]]) #获取(1,2)和(2,3)的两个数据，这也叫花式索引
```

```python
a1 = np.arange(0,24).reshape((4,6))
print(a1[:,2:4]) #获取0-1行的0-1列的数据
```

### 布尔索引：

```python
a1 = np.arange(0,24).reshape((4,6))
# print(a1<10)   #a<10
a1
```

```python
a1 = np.arange(0,24).reshape((4,6))
a2 = a1 < 10
print(a1[a2]) #这样就会在a1中把a2中为True的元素对应的位置的值提取出来
```
