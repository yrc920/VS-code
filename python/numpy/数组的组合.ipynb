### vstack

```python
import numpy as np
```

```python
a1 = np.random.randint(0,10,size=(3,5))
a2 = np.random.randint(0,10,size=(1,5))
```

```python
a1
```

```python
a2
```

```python
a3 = np.vstack([a1,a2])
a3
```

### hstack

```python
a1 = np.random.randint(0,10,size=(3,2))
a2 = np.random.randint(0,10,size=(3,1))
```

```python
a1
```

```python
a2
```

```python
a3 = np.hstack([a1,a2])
```

### concatenate([],axis)

```python
a = np.array([[1, 2],
              [3, 4]])  #(2,2)

b = np.array([[5, 6]]).T         #(1,2)
print(b)
np.concatenate((a, b), axis=1)
```
