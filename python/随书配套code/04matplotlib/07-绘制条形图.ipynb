```python
# 指定画布的尺寸大小
plt.figure(figsize=(20,5))
# for index in movie_df.index:
#     series = movie_df.iloc[index]
#     plt.bar(series.index,series.values)
# 每个电影之间需要保持一定距离，因此使用步长为 1.5 的间隔
xticks = np.arange(0,len(movies)*1.5,1.5)
bar_width = 0.2
# -2 -1 0 1 2
# plt.bar(xticks-2*bar_width,movie_df.iloc[0],width=bar_width)
# plt.bar(xticks-bar_width,movie_df.iloc[1],width=bar_width)
# plt.bar(xticks,movie_df.iloc[2],width=bar_width)
# plt.bar(xticks+bar_width,movie_df.iloc[3],width=bar_width)
# plt.bar(xticks+2*bar_width,movie_df.iloc[4],width=bar_width)
# 定义变量 bar_width 用于设置柱子的宽度，以及变量 index 迭代数据框的行索引。然后使用 plt.bar() 函数绘制柱状图，参数说明如下：
# x：表示每个柱子的横坐标位置，可以通过前面定义的 xticks 来设置，同时为了让多个柱子错开，还需要乘上一个系数；
# height：表示每个柱子的高度值，可以通过 pandas 的 iloc() 函数来选取数据框中的数值；
# width：表示每个柱子的宽度值，即前面定义的 bar_width 变量；
# label：表示每根柱子的标签，用于生成图例。
for index in movie_df.index:
    plt.bar(xticks+(-2+index)*bar_width,movie_df.iloc[index],width=bar_width,label='第%d天票房'%(index+1))
plt.xticks(xticks,movie_df.columns,fontproperties=font,size=12)
font.set_size(12)
plt.legend(prop=font)
plt.grid()
plt.show()
```

# 笔记：
1. 绘制条形图使用的是plt.bar方法。这个方法有很多参数：
    * x：在x轴上的坐标，可以为一个列表/数组/字符串。
    * y：在y轴上的坐标，可以为一个列表/数组/字符串。
    * data：如果传递了data，那么x和y就可以为data中的key。比如data是一个DataFrame对象，那么x和y就是这个DataFrame对象的某个列的名字。
2. 官方文档：https://matplotlib.org/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar
3. 横向条形图：plt.barh。然后y代表的是之前bar的x，width代表的是bar的height。其他参数都差不多。
4. 分组条形图：matplotlib并没有提供专门的分组条形图的API来实现。而是需要设置这个坐标来实现。绘制的规则是，一次性绘制所有电影第1天的数据，第二天的数据2...，然后要找到规律，把这个条形图的位置算好。
5. 堆叠条形图：matplotlib并没有提供专门的堆叠条形图的API来实现，我们在绘制往上堆叠的数据的时候，可以使用bottom参数，来往上偏移指定的距离。这样就可以实现堆叠的效果了。
