```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [np.random.randint(0,10) for x in range(10)]
# 抗锯齿技术是一种图像处理技术，用于减少线条和边缘产生的锯齿感。在 matplotlib 中，如果不启用抗锯齿（即 aa=False），
# 则图像中的线条和边缘可能会出现锯齿状，在高分辨率显示器上尤为明显。
plt.plot(y,lw=4,c='r',ls='--',aa=True)
```

```python
y1 = [np.random.randint(0,10) for x in range(10)]
y2 = [np.random.randint(0,10) for x in range(10)]
lines = plt.plot(range(10),y1,range(10),y2)
line = lines[0]
line.set_color('r')
# line1 = lines[1]
# line1.set_color('g')
line.set_linewidth(4)  # 设置线宽
line.set_alpha(0.5)  # #设置0.5的透明度
```

```python
y1 = [np.random.randint(0,10) for x in range(10)]
y2 = [np.random.randint(0,10) for x in range(10)]
lines = plt.plot(range(10),y1,range(10),y2)
plt.setp(lines,linewidth=4)
```

# 笔记：
1. 使用plot绘图的时候，就可以传递Line2D的属性值进去，来修改线条的样式。
2. 使用plot返回的线条，单独去设置也可以，但是单独设置的时候，是通过`set_Line2D属性名()`的方式来设置的。
3. 使用plt.setp来一次性设置多个线条的样式。
