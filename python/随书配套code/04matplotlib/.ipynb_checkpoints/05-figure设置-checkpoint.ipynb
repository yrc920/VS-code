```python
import matplotlib.pyplot as plt
import numpy as np
```

使用 plt.figure() 函数创建一个新的图形窗口，并指定一些参数来设置图形窗口的属性。具体参数说明如下：

figsize：用元组 (width, height) 指定图形窗口的尺寸，单位是英寸（inch）。

facecolor：指定图形窗口的背景色，可以是颜色名称或 RGB 值等格式。

edgecolor：指定图形窗口的边框颜色。

linewidth：指定图形窗口的边框宽度，单位是像素（pixel）。

dpi：指定图形窗口的分辨率（每英寸包含多少像素），默认值为 100。

frameon：一个布尔值，表示是否显示边框，默认值为 True。

```python
plt.figure(figsize=(10,5),facecolor="r",edgecolor="k",linewidth=2,dpi=200,frameon=True) # Axes
plt.plot(np.arange(5,20))
plt.savefig("x.png")
# ax = plt.gca()
# ax.set_facecolor("y")
```

# 笔记：
1. 画板的样式：通过调用plt.figure来实现，这个函数调用必须要在所有绘图之前来完成。参数解释：
    * figsize：画板的尺寸，是一个元组，第一个参数是宽度，第二个参数是高度。单位是英寸。
    * facecolor：画板颜色，注意区分Axes的背景颜色。
    * edgecolor：边框颜色，边框默认的宽度是0，因此如果想要看到边框，那么需要设置linewidth=2。
    * dpi：像素，也就是一个英寸中的像素点，默认是100，如果想要增加分辨率，那么就可以把dpi设置得大一点。
2. 保存图片：
    * plt.savefig。
    * 鼠标左键点击图片，然后右键保存图片。
