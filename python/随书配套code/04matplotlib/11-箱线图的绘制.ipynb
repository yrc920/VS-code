```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import font_manager
```

```python
# 生成一个包含 100 个随机数的 NumPy 数组 data，每个数在 0-100 之间
data = np.random.rand(100)*100
# 使用 NumPy 库中的 np.append() 函数，在数组 data 中追加了两个元素，分别为 -120 和 150。添加这两个元素会使得数据集中出现离群值。
data = np.append(data,np.array([-120,150]))
# 第一个参数 data 是数据集；sym='^' 代表使用三角形符号来表示离群值；vert=False 代表将箱线图设置为水平方向显示；
# widths=[0.1] 代表箱线的宽度为 0.1；labels=['A'] 代表该箱线图的标签为 A；meanline=True 代表显示平均线；showmeans=True 代表将平均值标出。
plt.boxplot(data,sym='^',vert=False,widths=[0.1],labels=['A'],meanline=True,showmeans=True)
```

# 笔记：
1. 箱线图（Box-plot）又称为盒须图、盒式图或箱型图，是一种用作显示一组数据分散情况资料的统计图。因形状如箱子而得名。在各种领域也经常被使用，它主要用于反映原始数据分布的特征，还可以进行多组数据分布特征的比较。箱线图的绘制方法是：先找出一组数据的上限值、下限值、中位数（Q2）和下四分位数（Q1）以及上四分位数（Q3）；然后，连接两个四分位数画出箱子；再将最大值和最小值与箱子相连接，中位数在箱子中间。
2. 使用matplotlib.pyplot.boxplot来绘制，这个上面有一些常用的参数：
    1. `x`：需要绘制的箱线图的数据。
    2. `notch`：是否展示置信区间，默认是`False`。如果设置为`True`，那么就会在盒子上展示一个缺口。
    3. `sym`：代表异常点的符号表示，默认是小圆点。
    4. `vert`：是否是垂直的，默认是`True`，如果设置为`False`那么将水平方向展示。
    5. `whis`：上下限的系数，默认是`1.5`，也就是上限是`Q3+1.5IQR`，可以改成其他的。也可以为一个序列，如果是序列，那么序列中的两个值分别代表的就是下限和上限的值，而不是再需要通过`IQR`来计算。
    6. `positions`：设置每个盒子的位置。
    7. `widths`：设置每个盒子的宽度。
    8. `labels`：每个盒子的`label`。
    9. `meanline`和`showmeans`：如果这两个都为`True`，那么将会绘制平均值的的线条。
