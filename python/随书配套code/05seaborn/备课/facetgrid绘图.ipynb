```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib as mpl
```

```python
tips = sns.load_dataset("tips")
sns.set(font="FangSong")
g = sns.FacetGrid(tips,col="day",row="time",subplot_kws={"facecolor":"g"},hue_kws={"marker":"^"})
g.map(sns.regplot,"total_bill","tip")
```

```python
axes = sns.scatterplot(x="total_bill",y="tip",data=tips)
axes.set_xticks(range(0,60,5))
```

```python
fig,[ax1,ax2] = plt.subplots(2,1,figsize=(10,10))
sns.scatterplot(x="total_bill",y="tip",data=tips,ax=ax1,hue="time")
sns.barplot(x="day",y="total_bill",data=tips,ax=ax2)
```
