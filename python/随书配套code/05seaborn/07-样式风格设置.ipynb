```python
# sns.set_style("white")
# sns.set_style("whitegrid")
# sns.set_style("dark")
# sns.set_style("darkgrid")
sns.set_style("ticks")
axes = sns.scatterplot(x="total_bill",y="tip",data=tips)
```
