```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
planets = sns.load_dataset("planets")
```

```python
# help(sns.barplot)
plt.figure(figsize=(20,5))
sns.barplot(x="year",y="distance",data=planets,ci=None,palette=sns.color_palette("hls",n_colors=23))
```

```python
current_palette = sns.color_palette("bright",n_colors=20)
sns.palplot(current_palette)
```

```python
sns.palplot(sns.color_palette("hls",100))
```

```python
sns.palplot(sns.hls_palette(n_colors=10,h=0.1,l=0.4,s=0.9))
```

```python
sns.palplot(sns.color_palette('Paired'))
```

```python
my_palette = ["mauve",'beige','salmon','peach','navy']
sns.palplot(sns.xkcd_palette(my_palette))
```
