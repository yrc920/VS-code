```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

在seaborn里最常用的观察单变量分布的函数是distplot(),默认地,这个函数会绘制一个直方图,并拟合一个核密度估计

```python
sns.set(color_codes=True)
the_titanic = titanic[~np.isnan(titanic['age'])]
# distplot() 函数，用于绘制 Titanic 号沉船事件中乘客的年龄分布图
# kde=True：表示是否绘制核密度估计曲线，即 KDE 线。默认值为 True，即绘制。
# bins=30：表示直方图分成的区间数，即直方图的柱子数目。默认值为 10，这里设置为 30。

# rug=True：表示是否在 X 轴上绘制小的短竖条，即数据点的位置。默认值为 False，这里设置为 True。

# hist=False：表示是否绘制直方图。默认值为 True，但由于我们使用了 KDE 曲线，因此将其设置为 False。
sns.distplot(the_titanic['age'],kde=True,bins=30,rug=True,hist=False)
```

jointplot() 函数则用于绘制两个变量之间的关系图，即用于展示两个变量的联合分布情况。它提供了多种绘图类型，例如散点图、KDE 曲线图和 Hexbin 图等，可以用于研究两个变量之间的相关性和分布情况。

```python
tips = sns.load_dataset("tips")
# sns.jointplot(x="total_bill",y="tip",data=tips)
# kind="reg"：表示要绘制的图形类型，即散点图加回归线。通过回归线，我们可以更好地展示账单金额和小费之间的趋势和相关性。
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")
```

这段代码使用 Seaborn 库的 jointplot() 函数，用于绘制中国运动员身高和体重之间的关系图。

具体来说，这个关系图是一个六边形图（hexbin），其中 X 轴表示身高（Height），Y 轴表示体重（Weight）。通过这个图形，我们可以看到身高和体重之间的分布情况，并探索这种关系的性质和强度。

代码中的参数含义如下：

x="Height"：表示要绘制的数据列，即中国运动员的身高数据列。

y="Weight"：表示要绘制的数据列，即中国运动员的体重数据列。

data=china_athletes：表示要绘制的数据集，即中国运动员身高和体重的数据集。这里使用了一个自定义的数据集 china_athletes，包括了 2016 年夏季奥运会上中国运动员的身高、体重、项目等信息。

kind="hex"：表示要绘制的图形类型，即六边形图。通过六边形图，我们可以更好地展示身高和体重之间的分布情况，并发现其中的规律和异常值。

color='r'：表示使用红色作为六边形图的颜色。

ratio=4：表示六边形图的宽高比，即横向和纵向的六边形数量比例。

space=0：表示横向和纵向的六边形之间的间距。

marginal_kws={"rug":True,'kde':True}：表示在图形边界上加入坐标轴和核密度估计图。通过这些信息，我们可以更好地理解身高和体重之间的分布情况和相关性。其中，rug=True 表示在坐标轴上画出小刻度线，kde=True 表示在边缘上画出核密度估计曲线。

```python
athletes = pd.read_csv("athlete_events.csv")
china_athletes = athletes[athletes['NOC']=='CHN']
sns.jointplot(x="Height",y="Weight",data=china_athletes,kind="hex",color='r',ratio=4,space=0,marginal_kws={"rug":True,'kde':True})
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

pairplot成对绘图：pairplot 可以把某个数据集中某几个字段之间的关系图一次性绘制出来。

```python
sns.pairplot(iris)
```

```python
sns.pairplot(tips,vars=['total_bill','tip'])

# 对角线上是各个属性的直方图（分布图），而非对角线上是两个不同属性之间的相关图。
```

kind="reg"：表示要绘制的图形类型，即散点图加回归线。通过回归线，我们可以更好地展示特征之间的趋势和相关性。

diag_kind="kde"：表示对角线上的小图绘制核密度估计曲线。通过这些曲线，我们可以更好地理解每个特征在数据集中的分布情况。

因此，这段代码将生成一个散点图矩阵，用于表示鸢尾花三个特征之间的关系。图形包括：

散点图和回归线：表示每对特征之间的散点图和线性回归模型拟合得到的关系线。通过这些图形，我们可以看到特征之间是否存在某种线性关系，并得到这个关系的具体形式。

对角线上的核密度估计曲线：表示每个特征在数据集中的分布情况。通过这些曲线，我们可以更好地理解每个特征的取值范围和分布情况。

```python
sns.pairplot(iris,vars=['sepal_length','sepal_width','petal_length'],kind="reg",diag_kind="kde")
```

# 分布绘图：
1. 单一变量绘图：用的是`distplot`，这个函数不仅仅可以绘制直方图，还可以绘制KDE曲线以及rug线，rug线约集中，说明数据约集中。
2. 多变量绘图：用的是`jointplot`，这个函数比传统的散点图可以展示更多的信息，他是在顶部和右边展示两个直方图。
3. 结对绘图（pairplot）：用的是`pairplot`，可以通过`vars`参数用来控制只需要绘制数据集中的某几个字段。斜对角的图，表示的是数据集中，单一列的分布情况。
