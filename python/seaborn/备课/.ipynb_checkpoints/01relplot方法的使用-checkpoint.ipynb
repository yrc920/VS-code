```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sns.set(style="ticks")
```

```python
tips = sns.load_dataset("tips",cache=True)
tips.head()
```

```python
sns.relplot(x="total_bill",y="tip",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",col="day",col_wrap=2,data=tips)
```
